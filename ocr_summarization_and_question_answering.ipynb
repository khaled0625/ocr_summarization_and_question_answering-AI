```python
from transformers import AutoModel, AutoTokenizer
#ocr model using GPU faster
tokenizer = AutoTokenizer.from_pretrained('ucaslcl/GOT-OCR2_0', trust_remote_code=True)
model = AutoModel.from_pretrained('ucaslcl/GOT-OCR2_0', trust_remote_code=True, low_cpu_mem_usage=True, device_map='cuda', use_safetensors=True, pad_token_id=tokenizer.eos_token_id)
model = model.eval().cuda()


```

```python
from transformers import  pipeline
# summarization model in english
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python

def summarize(text):
    # use summarization
    summary = summarizer(text, max_length=200, min_length=30, do_sample=False)
    return summary[0]['summary_text']
```

```python
from transformers import AutoModelForQuestionAnswering
# question-answering model in english

model_name = "deepset/roberta-base-squad2"

# a) Get predictions
pipe_qa_en = pipeline('question-answering', model=model_name, tokenizer=model_name)

# b) Load model & tokenizer
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python

# question-answering model in arabic
pipe_qa_ar = pipeline("question-answering", model="ZeyadAhmed/AraElectra-Arabic-SQuADv2-QA")
```

```python
# translation english to arabic
pipe_to_arabic = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ar")
```

```python



def to_arabic(text, max_length=512):
    # Split the text into chunks if it is more than 512
    chunks = [text[i:i+max_length] for i in range(0, len(text), max_length)]

    translated_chunks = []
    for chunk in chunks:
        translated_chunk = pipe_to_arabic(chunk, max_length=max_length)[0]['translation_text']
        translated_chunks.append(translated_chunk)

    # Join the translated chunks
    return ' '.join(translated_chunks)

```

```python
def question_answering(question,context):
    QA_input = {
        'question': question,
        'context': context,
    }
    res = pipe_qa_en(QA_input)
    return res['answer']


```

```python
def question_answering_ar(question,context):
    QA_input = {
        'question': question,
        'context': context,
    }
    res = pipe_qa_ar(QA_input)
    return res['answer']


```

```python
def process_image_and_text(image, text):



    ocr_text = model.chat(tokenizer, image, ocr_type='ocr')


    # Summarize
    res = summarize(ocr_text)

    # Combine with input text
    final_output = f"Input text: {text}\n\nProcessed OCR text: {res}"

    return final_output

```

```python
def process_image_qa(language,image, text):



    ocr_text = model.chat(tokenizer, image, ocr_type='ocr')


    #if arabic selected
    if language == 'arabic':
      arabic_context = to_arabic(ocr_text) # translation to arabic
      arabic_answer = question_answering_ar(text,arabic_context)
      return arabic_answer



    # Summarize and fix spelling mistakes
    res = question_answering(text,ocr_text)

    # Combine with input text
    final_output = f"Input text: {text}\n\nthe answer is : {res}"

    return final_output
```

```python
import gradio as gr
from transformers import AutoModelForCausalLM, AutoProcessor, GenerationConfig
from PIL import Image


# Create Gradio interface
summarization_Interface = gr.Interface(
    fn=process_image_and_text,
    inputs=[
        gr.Image(type="filepath", label="Upload Image"),
        gr.Textbox(label="Input Text", placeholder="Type your text here...")
    ],
    outputs=gr.Textbox(label="output Text"),
    title="Image and Text Processor",
    description="Upload an image and provide some text to generate a response."
)


qa_Interface =  gr.Interface(
    fn=process_image_qa,
    inputs=[
        gr.Radio(['Arabic', 'English'], label='Select Language',value= 'Arabic'),
        gr.Image(type="filepath", label="Upload Image"),
        gr.Textbox(label="Input question", placeholder="Type your question here...")
    ],
    outputs=gr.Textbox(label="Answer Text"),
    title="Image and Text Processor",
    description="Upload an image and provide some text to generate a response."
)

apps_interface = gr.TabbedInterface(
    [summarization_Interface, qa_Interface],
    tab_names=["summarization", "question answering"]
)

# Launch the interface
apps_interface.launch()
```
